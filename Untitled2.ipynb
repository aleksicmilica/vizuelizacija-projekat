```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv')

app = Dash(__name__)

app.layout = html.Div([
    html.H1(children='Title of Dash App', style={'textAlign':'center'}),
    dcc.Dropdown(df.country.unique(), 'Canada', id='dropdown-selection'),
    dcc.Graph(id='graph-content'),
    dcc.Dropdown(df.country.unique(), 'Canada', id='dropdown-selection2'),
    dcc.Graph(id='graph-content2'),
    html.H1(children='Customer Segmentation Analysis', style={'textAlign': 'center'}),

    html.Div([
        dcc.Graph(id='pie-chart'),
        dcc.Graph(id='bar-chart'),
        dcc.Graph(id='treemap'),
        dcc.Graph(id='stacked-bar-chart')
    ])
])

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')

)
def update_graph(value):
    dff = df[df.country==value]
    return px.line(dff, x='year', y='pop')
@callback(
    Output('graph-content2', 'figure'),
    Input('dropdown-selection2', 'value')

)
def update_graph2(value):
    dff = df[df.country==value]
    return px.line(dff, x='year', y='pop')

if __name__ == '__main__':
    app.run(debug=True)
```

```python
recnik_kategorije={
    'Furniture':"Namestaj",
    'Office Supplies':"Kancelarijski pribor",
    'Technology':"Tehnologija"
} 
```

```python
data.replace({"Category":recnik_kategorije},inplace=True)
```

```python
data["Region"].unique()
```

```python
recnik_regioni={
    'South':"Jug",
    'West':"Zapad",
    'Central':"Centar",
    "East":"Istok"
} 
data.replace({"Region":recnik_regioni},inplace=True)
```

```python
data["Segment"].unique()
```

```python
recnik_segmenti={
    'Consumer':"Potrosac",
    'Corporate':"Korporacija",
    'Home Office':"Kucna kancelarija",
    
} 
data.replace({"Segment":recnik_segmenti},inplace=True)
```

```python
data['Order Date'] = data['Order Date'].astype(str)  
data[["Month","Day","Year"]] = data['Order Date'].str.split('/',expand=True).astype(int)
data["Quartile"] = data["Month"].astype(int)//4+1
sales_per_month = data.groupby(["Month"])[["Sales","Discount","Quantity"]].sum().sort_index(ascending=True)
sales_per_quartile = data.groupby(["Quartile"])[["Sales","Discount","Quantity"]].sum().sort_index(ascending=True)
sales_per_year = data.groupby(["Year"])[["Sales","Discount","Quantity"]].sum().sort_index(ascending=True)
```

```python
sales_per_month.head()
```

```python
sales_per_category_month = data.groupby(["Category","Month"])["Sales"].sum()
sales_per_category_quartile = data.groupby(["Category","Quartile"])["Sales"].sum()
sales_per_category_year = data.groupby(["Category","Year"])["Sales"].sum()
# sales_per_category_year
```

```python
sales_per_region_month = data.groupby(["Region","Month"])["Sales"].sum()
sales_per_region_quartile = data.groupby(["Region","Quartile"])["Sales"].sum()
sales_per_region_year = data.groupby(["Region","Year"])["Sales"].sum()
```

```python
impact_factor = {
    "Discount":"Uticaj popusta",
    "Quantity":"Uticaj kolicine"
}
for factor, impact in impact_factor.items():
    sales_per_month[impact] = sales_per_month["Sales"].diff()-sales_per_month[factor]
sales_per_month.fillna(0,inplace=True)
```

```python
sales_per_month.reset_index()[["Month"]]
```

```python
df = sales_per_year.reset_index()
df_temp = df[["Sales","Discount","Quantity"]]
# df_temp.columns = ["mesec", 'Prodaja',"Popust","Kolicina"]
df
```

```python
app = Dash(__name__)

sales_period = ["mesec","kvartal","godina"]
########################################
sales = sales_per_month['Sales']
months = sales.index
discount_impact = sales_per_month['Uticaj popusta']
quantity_impact = sales_per_month['Uticaj kolicine']

# # Calculate cumulative impacts
# cumulative_discount_impact = discount_impact.cumsum()
# cumulative_quantity_impact = quantity_impact.cumsum()

graph_sales_impact = px.bar(sales,x=months, y=sales)

graph_discount_impact = px.bar(discount_impact,x=months, y=discount_impact, color='Uticaj popusta',  color_continuous_scale='RdYlBu')
graph_sales_impact.add_trace(graph_discount_impact.data[0])

graph_quantity_impact = px.bar(quantity_impact,x=months, y=quantity_impact, color='Uticaj kolicine', color_continuous_scale='Viridis')
graph_sales_impact.add_traces(graph_quantity_impact.data[0])
graph_sales_impact.update_layout(
    xaxis_title='Mesec',  # Title for x-axis
    yaxis_title='Prodaja',   # Title for y-axis
    # You can add more layout customization options here
)
#####################################
segment_counts = data['Segment'].value_counts()
pic_chart_customer_segment = px.pie(names=segment_counts.index, values=segment_counts.values, title='Raspodela kupaca po segmentima')


# order_value = data.groupby('Segment')['Sales']
# df_order_value = sales_per_year.reset_index()
# df_order_value.columns = ["Segment","Sales"]
# boxplot_customer = px.box(df_order_value,y='Sales', x='Segment',  title='Distribucija narucivanja po segmentima')
# boxplot_customer.update_layout(
#     yaxis_title='Prodaja',
# )

customer_counts = data.groupby(['Region', 'Segment']).size().reset_index(name='Customer Count')
treemap_customer_distr = px.treemap(customer_counts, path=['Region', 'Segment'], values='Customer Count', title='Distribucija potrosaca po regionima i segmentima')


customer_counts = data.groupby(['Segment', 'Category']).size().reset_index(name='Customer Count')
bar_segment_cat = px.bar(customer_counts, x='Segment', y='Customer Count', color='Category', barmode='stack', title='Raspodela segmenta potrosaca po kategoriji prozivoda')

###############################################
# data['Order Date'] = pd.to_datetime(data['Order Date'])
# data.sort_values("Sales",inplace=True)
# line_top_salled_product = px.line(data, x='Order Date', y='Sales', color='Product Name', title='Sales trends of the top-selling products over time')

# bubble_rel_product = px.scatter(data, x='Quantity', y='Profit', size='Discount', color='Product Name',
#                      title='Relationship between product sales volume, profit margin, and discount amount')

boxplot_dist_sales = fig = px.box(data, y='Sales',x='Category',log_y=True , title='Distribution of sales prices for each product category')
radar_data = data.groupby('Product Name')[['Sales', 'Profit']].mean().reset_index()
# radar_comp_product = px.line_polar(radar_data, r=['Sales', 'Profit'], theta='Product Name', line_close=True,
#                         title='Comparison of product performance metrics across different products')
    

###############################################

app.layout = html.Div([
    html.H1(children='Analiza poslovanja', style={'textAlign':'center'}),
    dcc.Dropdown(sales_period, 'mesec', id='dropdown-selection-sales-period'),
    dcc.Graph(id='graph-sales-period'),
    dcc.Dropdown(sales_period, 'mesec', id='dropdown-selection-sales-period-category'),
    dcc.Graph(id='graph-sales-period-category'),
    dcc.Dropdown(sales_period, 'mesec', id='dropdown-selection-sales-period-region'),
    dcc.Graph(id='graph-sales-period-region'),
    dcc.Graph(id='graph-sales-impact',figure = graph_sales_impact),
    html.H1(children='Customer Segmentation Analysis', style={'textAlign': 'center'}),

    html.Div([
        dcc.Graph(id='pie-chart',figure=pic_chart_customer_segment),
#         dcc.Graph(id='boxplot_customer',figure = boxplot_customer),
        dcc.Graph(id='treemap',figure=treemap_customer_distr),
        dcc.Graph(id='stacked-bar-chart',figure = bar_segment_cat)
    ]),
    html.H1(children='Sales Analysis', style={'textAlign': 'center'}),

    html.Div([
#         dcc.Graph(id='line_top_salled_product',figure = line_top_salled_product),
#         dcc.Graph(id='bubble_rel_product',figure = bubble_rel_product),
        dcc.Graph(id='boxplot_dist_sales',figure = boxplot_dist_sales),
#         dcc.Graph(id='radar_comp_product',figure = radar_comp_product)
    ])
#     dcc.Dropdown(df.country.unique(), 'Canada', id='dropdown-selection2'),
#     dcc.Graph(id='graph-content2')
])


#######################
@callback(
    Output('graph-sales-period', 'figure'),
    [Input('dropdown-selection-sales-period', 'value')]

)
def update_bar_sales_period(value):
    
    if value == "mesec":
        df = sales_per_month.reset_index()
        df_temp = df[["Month","Sales","Discount","Quantity"]]
    elif value == "kvartal":
        df = sales_per_quartile.reset_index()
        df_temp = df[["Quartile","Sales","Discount","Quantity"]]
    else:
        
        df = sales_per_year.reset_index()
        df_temp = df[["Year","Sales","Discount","Quantity"]]
    df_temp.columns = [value, 'Prodaja',"Popust","Kolicina"]
    tick_labels = df_temp[value].unique()

    
    fig = px.bar(df_temp, x=value, y='Prodaja')
    fig.update_layout(xaxis=dict(tickvals=tick_labels, ticktext=tick_labels))
    
    return fig
##########################
@callback(
    Output('graph-sales-period-category', 'figure'),
    [Input('dropdown-selection-sales-period-category', 'value')]

)
def update_bar_sales_period_category(value):
    if value == "mesec":
        df = sales_per_category_month.reset_index()
    elif value == "kvartal":
        df = sales_per_category_quartile.reset_index()
    else:
        
        df = sales_per_category_year.reset_index()
    df.columns = ["Kategorija",value,"Prodaja"]
    tick_labels = df[value].unique()
    fig = px.bar(df, x=value, y="Prodaja", color="Kategorija")
    fig.update_layout(xaxis=dict(tickvals=tick_labels, ticktext=tick_labels))
    
    return fig
#########################################
@callback(
    Output('graph-sales-period-region', 'figure'),
    [Input('dropdown-selection-sales-period-region', 'value')]

)
def update_bar_sales_period_region(value):
    if value == "mesec":
        df = sales_per_region_month.reset_index()
    elif value == "kvartal":
        df = sales_per_region_quartile.reset_index()
    else:
        
        df = sales_per_region_year.reset_index()
    df.columns = ["Region",value,"Prodaja"]
    
    
    tick_labels = df[value].unique()
    fig = px.area(df, x=value, y="Prodaja", color="Region")
    fig.update_layout(xaxis=dict(tickvals=tick_labels, ticktext=tick_labels))
    
    return fig
##############################################

###########################################
if __name__ == '__main__':
    app.run(debug=True)
```
