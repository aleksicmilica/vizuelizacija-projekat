```python
import pandas as pd
import panel as pn
import matplotlib
import matplotlib.pyplot as plt

pn.extension(design='material')
pn.extension('ipywidgets')
plt.switch_backend('agg')
```

```python
url = 'https://raw.githubusercontent.com/aleksicmilica/vizuelizacija-projekat/main/WalmartSQL%20repository..csv'
dataset = pd.read_csv(url,delimiter=";")
```

```python
revenue_per_period = dataset.groupby(["time_of_day"])["total"].sum()
revenue_per_day = dataset.groupby(["day_name"])["total"].sum()
revenue_per_month = dataset.groupby(["month_name"])["total"].sum()
```

```python
revenue_per_period.values
```

```python
def create_graph(seria,xlabel,ylabel,title):

       fig, ax = plt.subplots()
       ax.plot(seria.index,seria.values)

       ax.set(xlabel=xlabel, ylabel=ylabel,
              title=title)
       ax.grid()

       plt.close(fig)
       return fig

```

```python
rev_per_period_graph = pn.pane.Matplotlib(create_graph(revenue_per_period,"period","revenue","revenue per period"))
```

```python
def on_select_period_change(event):
  if matplotlib.get_backend() == 'agg':
    tip = select_graph_period.value
    if tip == "period dana":
      rev_per_period_graph.object = create_graph(revenue_per_period,"period","prihod","prihod po periodu")
    else:
      if tip == "dan u nedelji":
        rev_per_period_graph.object = create_graph(revenue_per_day,"dan","prihod","prihod po danu")
      else:
        rev_per_period_graph.object = create_graph(revenue_per_month,"mesec","prihod","prihod po mesecu")
    print(tip)
```

```python
list_graph_period = ["period dana","dan u nedelji", "mesec"]
select_graph_period = pn.widgets.Select(value="period dana",options=list(list_graph_period))
select_graph_period.param.watch(on_select_period_change,"value")
```
